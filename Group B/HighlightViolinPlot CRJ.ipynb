```python
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import plotly.io as pio

# Load the dataset
file_path = '/Users/wangzhuoyulucas/Library/Mobile Documents/com~apple~CloudDocs/ALLABOUTDSS/DSS5105/5105-Question2/merged_df1.csv'  # Replace with the correct file path
df = pd.read_csv(file_path)

# Plot: Interactive Histogram of GHG Emissions for All Buildings and Highlighted Building
fig_histogram = go.Figure()

# Plot the overall distribution in grey as the background
fig_histogram.add_trace(go.Histogram(x=df['GHG_Total'], nbinsx=20, name='All Buildings', marker_color='grey', opacity=0.3))

# Highlight specific building(s) with different colors
highlight_buildings = ['EQUINIX SG3 DATA CENTRE', 'BLK 20 ARC']
colors = ['red', 'blue']
for building, color in zip(highlight_buildings, colors):
    building_ghg = df[df['Building Name'] == building]['GHG_Total']
    fig_histogram.add_trace(go.Histogram(x=building_ghg, nbinsx=20, name=building, marker_color=color, opacity=0.7))

fig_histogram.update_layout(title='Interactive Histogram of GHG Emissions with Highlighted Buildings',
                            xaxis_title='Total GHG Emissions (ton)',
                            yaxis_title='Count',
                            barmode='overlay')

# Show the plot
pio.show(fig_histogram)
```

```python
import pandas as pd
import plotly.graph_objs as go
import plotly.express as px
import plotly.io as pio
import numpy as np

# Load the dataset
file_path = '/Users/wangzhuoyulucas/Library/Mobile Documents/com~apple~CloudDocs/ALLABOUTDSS/DSS5105/5105-Question2/merged_df1.csv'  # Replace with the correct file path
df = pd.read_csv(file_path)

# Apply log transformation to reduce skewness
# Adding a small constant to avoid log(0) if needed
df['GHG_Total_Log'] = np.log(df['GHG_Total'] + 1)

# Calculate percentiles
percentiles = np.percentile(df['GHG_Total_Log'], [25, 50, 75])

# Create a figure for the swarm plot and violin plot
fig = go.Figure()

# Add a violin plot for GHG emissions by building (log-transformed)
fig.add_trace(go.Violin(
    x=df['GHG_Total_Log'],
    name='GHG Emissions (Log Transformed)',
    box_visible=True,
    meanline_visible=True,
    fillcolor='lightblue',
    line_color='darkblue',
    opacity=0.6,
    orientation='h'
))

# Add swarm points to the plot (log-transformed) for non-highlighted buildings in grey
non_highlight_df = df[~df['Building Name'].isin(['EQUINIX SG3 DATA CENTRE', 'BLK 20 ARC'])]
fig.add_trace(go.Scatter(
    y=['GHG Emissions (Log Transformed)'] * len(non_highlight_df),
    x=non_highlight_df['GHG_Total_Log'],
    mode='markers',
    marker=dict(size=6, color='grey', symbol='circle'),
    name='All Buildings (Non-Highlighted)'
))

# Highlight specific building(s) with larger marker size and dodged position (log-transformed)
highlight_buildings = ['EQUINIX SG3 DATA CENTRE', 'BLK 20 ARC']
highlight_df = df[df['Building Name'].isin(highlight_buildings)]
fig.add_trace(go.Scatter(
    y=['GHG Emissions (Log Transformed)'] * len(highlight_df),
    x=highlight_df['GHG_Total_Log'],
    mode='markers',
    marker=dict(size=14, color='orange', symbol='diamond'),
    name='Highlighted Buildings'
))

# Add percentile lines
for percentile, label in zip(percentiles, ['25th Percentile', '50th Percentile', '75th Percentile']):
    fig.add_trace(go.Scatter(x=[percentile, percentile], y=[0.8, 1.2],
                             mode='lines',
                             line=dict(color='red', dash='dash'),
                             name=label))

fig.update_layout(title='Violin Plot with Swarm Points and Percentiles for GHG Emissions (Log Transformed)',
                  xaxis_title='Log of Total GHG Emissions (ton)',
                  yaxis_title='',
                  plot_bgcolor='rgba(240, 240, 240, 1)',
                  font=dict(size=14),
                  xaxis=dict(showgrid=True, gridcolor='lightgrey'),
                  yaxis=dict(showgrid=True, gridcolor='lightgrey'))

# Show the plot
pio.show(fig)
```
